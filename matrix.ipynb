```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
v = np.array([3, 2])
t = np.array([1, -2])
w = np.array([1, 4, 2])

print("v =", v)
print("t =", t)
print("w =", w)
```

```python
A = np.array([[2, 0],
              [1, 2]])

B = np.array([[3, 0],
              [0, -1]])

C = np.array([[1/2,  np.sqrt(3)/2],
              [-np.sqrt(3)/2, 1/2]])

print("Matrix A:\n", A)
print("\nMatrix B:\n", B)
print("\nMatrix C:\n", C)
```

```python
Av = A @ v
Bv = B @ v
Cv = C @ v
ACv = A @ Cv

print("Av =", Av)
print("Bv =", Bv)
print("Cv =", Cv)
print("ACv =", ACv)
```

```python
ACt = A @ C.T
CAt = C.T @ A

ABt = A @ B.T
BAt = B.T @ A

BCt = B @ C.T
CBt = C.T @ B

print("AC^T:\n", ACt)
print("\nC^T A:\n", CAt)

print("\nAB^T:\n", ABt)
print("\nB^T A:\n", BAt)

print("\nBC^T:\n", BCt)
print("\nC^T B:\n", CBt)
```

```python
print("Is AC^T equal to C^T A ?", np.allclose(ACt, CAt))
print("Is AB^T equal to B^T A ?", np.allclose(ABt, BAt))
print("Is BC^T equal to C^T B ?", np.allclose(BCt, CBt))
```

```python
plt.figure()

origin = [0, 0]

plt.quiver(*origin, v[0], v[1], angles='xy', scale_units='xy', scale=1)
plt.quiver(*origin, Av[0], Av[1], angles='xy', scale_units='xy', scale=1)
plt.quiver(*origin, Bv[0], Bv[1], angles='xy', scale_units='xy', scale=1)
plt.quiver(*origin, Cv[0], Cv[1], angles='xy', scale_units='xy', scale=1)

plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.axhline(0)
plt.axvline(0)

plt.gca().set_aspect('equal', adjustable='box')
plt.grid()
plt.title("Vector Transformations")
plt.show()
```

```python
D = np.array([[-np.sqrt(2)/2, 0,  np.sqrt(2)/2],
              [0, 1, 0],
              [np.sqrt(2)/2, 0, -np.sqrt(2)/2]])

print("Matrix D:\n", D)
```

```python
Dw = D @ w

print("Dw =", Dw)
```
